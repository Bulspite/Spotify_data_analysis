```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_tracks = dft.query('popularity > 90').sort_values('popularity',ascending = False)
top_tracks.head()
```

```python
tt = dft[dft['popularity']> 90].sort_values('popularity',ascending = False)
tt.head()
```

```python
dft["release_date"] = pd.to_datetime(dft["release_date"])
```

```python
dft  = dft.set_index('release_date')
dft.head()
```

```python
dft["duration"] = dft["duration_ms"].apply(lambda x: round(x/1000))
dft = dft.drop('duration_ms', axis = 1)
```

```python
corr_df = dft.drop(["key","mode","explicit"],axis =1).corr(method ="pearson")
plt.figure(figsize =(14,6))
hmap = sns.heatmap(corr_df,annot = True,vmax = 1,vmin = -1,cmap="inferno")
hmap.set_title("Correlation between Variables")
hmap.set_xticklabels(hmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = dft.sample(int(0.004*len(dft)))
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x ='energy', y ='loudness', line_kws={"color": "red"}).set(title = "Loudness vs Energy")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x ='popularity', y ='acousticness', line_kws={"color": "green"}).set(title = "Acousticness vs Popularity")
```

```python
dft['dates'] = dft.index.get_level_values('release_date')
dft.dates= pd.to_datetime(dft.dates)
years = dft.dates.dt.year
```

```python
sns.displot(years,height = 6, aspect = 3).set(title='Songs per Year')
```

```python
total_dr = dft.duration
fig , ax = plt.subplots(figsize = (18,7))
fig = sns.barplot(x = years, y = total_dr, ax = ax,errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation = 90)
```

```python
dfg =pd.read_csv("SpotifyFeatures.csv")
dfg.head()
```

```python
plt.title("Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre' , x = 'duration_ms' , data = dfg)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10,5))
fam = dfg.sort_values('popularity', ascending = False).head(10)
sns.barplot(x='popularity', y='genre', data= fam).set(title="Top Genres by Popularity")
```
